# The *new* **Network** Class

## Intro

This notebook demonstrates the new design of skrf's **Network** Class. The new class utilizes a more object-oriented approach which is cleaner and more scalable. The draw-back is that it breaks backward compatibility.

Creating a *new style* **Network** from an old

```python
from skrf import network2
reload(network2)
a = network2.Network.from_ntwkv1(rf.data.ring_slot)
```

The new **Network** class employs nested classes, which make for a cleaner and more logical namespace. The basic structure is:

* **Network**
    * **Frequency ** (same as before)
    * **Parameter** (s,z,y,etc)
        * **Projection** ( db10, db20, deg, mag,etc)
 
 

## Parameters

Accessing a Network's parameters returns on object, not an array as before

```python
type(a.s)
```

You can get at the array by accessing the property `val`.

```python
a.s.val[:3]
```

You can also slice the parameter directly. This can be used as an alternative way to access the values. 

```python
a.s[:2]
```

This nested-object desgin allows for simplified and more concise function calls as well as a more robust code-base. For example, plot functions become members of the parameters, which behave like you expect

```python
a.s.plot()
```

```python
a.s.plot_smith()
```

## Projections

A further advantage is that each parameter has members which represent various scalar projections. 

```python
type(a.s.db)
```

Their numerical values may be accessed through the `val` or by slicing, just like a parameter

```python
a.s.db[:3]
```

Projections also have plot as you expect

```python
a.s.db.plot();
```
